```
import pandas as pd
import sys
import time as t

sys.setrecursionlimit(10000)
pd.set_option('display.max_columns',None,'display.max_rows',None)

#-----------------SELECTIONSORT---------------------------

def selectionSort(data):
    inicio = t.time()
    comp = 0
    swap = 0
    data = data.to_numpy()
    for i in range (0, len(data) - 1):
        MIN = data[i].copy()
        MINP = i
        for j in range(i+1, len(data)):
            comp += 1
            if MIN[0] > data[j][0]:
                MIN = data[j].copy()
                MINP = j
        data[MINP], data[i] = data[i].copy(), MIN
        if MINP != i:
          swap += 1 
    fim = t.time()
    tempo = fim - inicio
    resultados(tempo, comp, swap)

#-----------------INSERTIONSORT---------------------------

def insertionSort(data): 
    inicio = t.time()
    comp = 0
    swap = 0
    data = data.to_numpy()
    for i in range(1,len(data)):
        key = data[i].copy()
        j = i - 1
        flag = False
        while j>=0 and key[0] < data[j][0]:
            flag=True
            comp += 1
            data[j+1] = data[j].copy()
            data[j] = key.copy()
            j = j - 1
            swap +=1
            comp +=1
        if flag == False:
            comp +=1
    fim = t.time()
    tempo = fim - inicio
    resultados(tempo, comp, swap)



#--------------------MERGESORT----------------------------

mergeComp = 0
mergeSwap = 0
def mergeSort(data, inicio = 0, fim = None):
        global mergeComp
        global mergeSwap
        if fim is None:
            fim = len(data)
            data = data.to_numpy()
        if (fim - inicio > 1):
            meio = (inicio + fim)//2
            mergeSort(data, inicio, meio)
            mergeSort(data, meio, fim)
            juntar(data, inicio, meio, fim)
        return {"comp": mergeComp, "swap": mergeSwap}

def juntar(data, inicio, meio, fim):
    global mergeSwap
    global mergeComp
    esquerda = data[inicio:meio].copy()
    direita = data[meio:fim].copy()
    i = 0
    j = 0
    for k in range(inicio, fim):
        if i >= len(esquerda):
            data[k] = direita[j].copy()
            j = j + 1
            mergeSwap += 1
        elif j >= len(direita):
            data[k] = esquerda[i].copy()
            i = i + 1
            mergeSwap += 1
        elif esquerda[i][0] < direita[j][0]:
            data[k] = esquerda[i].copy()
            i = i + 1
            mergeSwap += 1
        else:
            data[k] = direita[j].copy()
            j = j + 1
            mergeSwap += 1
        if i < len(esquerda) and j < len(direita):
          mergeComp += 1

#--------------------QUICKSORT----------------------------

quickComp = 0
quickSwap = 0
def quickSort(data, inicio = 0, fim= len(data) - 1, flag = True):
    global quickComp
    global quickSwap
    if flag is True:
        data = data.to_numpy()
        flag = False
    if inicio >= fim:
        return {"comp":quickComp, "swap":quickSwap}

    p = partition(data, inicio, fim)
    quickSort(data, inicio, p-1, flag)
    quickSort(data, p+1, fim, flag)   
    return {"comp":quickComp, "swap":quickSwap} 

def partition(data, inicio, fim):
    global quickComp
    global quickSwap
    pivo = data[inicio]
    menor = inicio + 1
    maior = fim

    while True:
        while menor <= maior and data[maior][0] >pivo[0]:
            maior = maior - 1
            quickComp+=1

        while menor <= maior and data[menor][0] <= pivo[0]:
            menor = menor + 1
            quickComp+=1

        if menor <= maior:
            data[menor], data[maior] = data[maior].copy(), data[menor].copy()
            quickSwap+=1

        else:
            break

    data[inicio], data[maior] = data[maior].copy(), data[inicio].copy()
    quickSwap+=1

    return maior

#--------------------Resultados----------------------------

def resultados(tempo, comp, trocas):
      print(f'Tempo para ordenação: {tempo}s',f'\nComparações: {comp}', f'\nTrocas: {trocas}')
```
